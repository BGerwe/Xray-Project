```python
direc=r'G:\My Drive\UW Research\fr-XAS November 2019\Data\fr-XAS\DPB2\700C_100% O2\2 Hz\150 mV'
# direc=r'C:\Users\shadd\OneDrive\Documents\Python\Xray-project\fr_XAS\November 2019 Test Data\191101_Amezawa\fr_XAS\TPB2\100% O2\fr-XAS\5 Hz\119 mV'

all_files = glob.glob(direc + '\*Xpress.txt')
all_files.sort()


fig, (ax1,ax2,ax3) = plt.subplots(ncols=1,nrows=3,figsize=(15,10))

Iravg = []
Iavg = []
Vavg = []


for i in range(len(all_files)-3):
    dat = pd.read_csv(all_files[i], delimiter='\t', header=0)
    
    
    
    Ir = np.array(dat['ROI1sum/I0'])
    t = np.array(dat['Time'])
    I = np.array(dat['Current'])
    V = np.array(dat['Voltage'])
    Pos = np.array(dat['Position'])
    
    if i == 0:
        Iravg = Ir
        Iavg = I
        Vavg = V
    else:
        Iravg = Iravg+Ir
        Iavg = Iavg+I
        Vavg = Vavg+V
        
    Iravg = Iravg/len(all_files)
    Iavg = Iavg/len(all_files)
    Vavg = Vavg/len(all_files)
    
    t_str = 20
    t_end = 40
#     ax1.plot(t,Ir)
    ax1.set_xlim(t_str,t_end)
    ax2.plot(t,V)
    ax2.set_xlim(t_str,t_end)
    ax3.plot(t,I)
    ax3.set_xlim(t_str,t_end)
    
ax1.plot(t, Iravg)
ax1.set_ylim(.095, .11)
plt.show()
```
