```python
def sortfunc(all_files):
    return int(all_files.split('\\P')[-1])

def sortfunc2(all_files):
    return int(all_files.split('mV ')[-1].split('_')[0][1:])

def sort_by_pos(data, pos):
    data_adj = [a for _,a in sorted(zip(pos, data))]
    return np.array(data_adj)
```
