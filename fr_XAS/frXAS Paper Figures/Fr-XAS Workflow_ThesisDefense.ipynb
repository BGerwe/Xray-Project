```python
read_direc = r'E:\\UW Research\\fr-XAS July 2018\\Data\\1% O2\\1 Hz 150 mV'
suffix = r'\\P3'
run_str = '\\R'
last_run = 1

match_str = suffix + run_str + f'{str(last_run)} [0-9]*.txt'

suffix_str = suffix.split('\\')[-1]
print(f'Analyzing {read_direc} {suffix_str} R{last_run}')

ti, v, j, ir, ir_avg = time_domain.extract_data(read_direc, match_str, start=0, end=None,
                     xray_disp=True)
step = 5
ir = -ir[::step]
ti = ti[::step]
v = time_domain.sub_mean(v[::step])
```

```python
freqs_0 = np.fft.fftshift(np.fft.fftfreq(len(ti), d=ti[1]))
f = 1

v_fft_0 = time_domain.get_windowed_fft(ti, v, f, 4)
v_fft_fit_0 = time_domain.fit_windowed_fft(freqs, v_fft, f, 4, harmonics=3)
ir_fft_0 = time_domain.get_windowed_fft(ti, ir, f, 4)
ir_fft_fit_0 = time_domain.fit_windowed_fft(freqs, ir_fft, f, 4, harmonics=3)
```

```python
font = {'family' : 'arial',
        'weight' : 'normal',
        'size'   : 10}
mpl.rc('font', **font)

mu_mult = 10
ms = 4
plot_step = 7
freqs = freqs_0[::plot_step]
v_fft = v_fft_0[::plot_step]
v_fft_fit_eval = v_fft_fit_0.eval()[::plot_step]
ir_fft = ir_fft_0[::plot_step]
ir_fft_fit_eval = ir_fft_fit_0.eval()[::plot_step]


_, axes = plt.subplots(nrows=2, figsize=(3.5,5))

axes[0].plot(freqs, v_fft.real, 'kx', ms=ms)
axes[0].plot(freqs, v_fft_fit_eval.real, 'k-', label ='V')
axes[0].plot(freqs, ir_fft.real* mu_mult, 'bo', ms=ms, fillstyle='none')
axes[0].plot(freqs, ir_fft_fit_eval.real* mu_mult, 'b--', label =f'$\chi$ x {mu_mult}')

axes[0].set_xticklabels([])
axes[0].set_ylabel('Re[FFT]', fontsize=14)
axes[0].legend()


axes[1].plot(freqs, v_fft.imag, 'kx', ms=ms)
axes[1].plot(freqs, v_fft_fit_eval.imag, 'k-', label ='V')
axes[1].plot(freqs, ir_fft.imag* mu_mult, 'bo', ms=ms, fillstyle='none')
axes[1].plot(freqs, ir_fft_fit_eval.imag* mu_mult, 'b--', label =f'$\chi$ x {mu_mult}')

axes[1].set_xlabel('Frequency (Hz)', fontsize=14)
axes[1].set_ylabel('Im[FFT]', fontsize=14)

for ax in axes:
    ax.set_xlim(0, 3.5)
    ax.set_ylim(-3e-4, 3e-4)
    ax.ticklabel_format(axis='y', scilimits=(0,0))

plt.subplots_adjust(hspace=0.05)
plt.savefig(r'G:\\My Drive\UW Research\Thesis\Presentation\Inkscape\FFT_1Hz_P3_V_Ir.svg', bbox_inches='tight', dpi=600)

plt.show()
```
