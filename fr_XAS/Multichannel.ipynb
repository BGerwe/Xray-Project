```python
direc=r'G:\My Drive\UW Research\fr-XAS November 2019\Data\TPB2\100% O2\fr-XAS\13 Hz\164 mV'

all_files=glob.glob(direc + '\*Xpress.txt')
all_files.sort()


fig, (ax1,ax2,ax3) = plt.subplots(ncols=1,nrows=3,figsize=(10,10))

Iravg=[]
Iavg=[]
Vavg=[]


for i in range(len(all_files)):
    dat = pd.read_csv(all_files[i], delimiter='\t', header=0)
    
    
    
    Ir=np.array(dat['ROI1sum/I0'])
    t=np.array(dat['Time'])
    I=np.array(dat['Current'])
    V=np.array(dat['Voltage'])
    Pos=np.array(dat['Position'])
    
    if i==0:
        Iravg=Ir
        Iavg=I
        Vavg=V
    else:
        Iravg=Iravg+Ir
        Iavg=Iavg+I
        Vavg=Vavg+V
        
    Iravg=Iravg/len(all_files)
    Iavg=Iavg/len(all_files)
    Vavg=Vavg/len(all_files)
    
    
    ax1.plot(Pos,Ir)
    ax1.set_xlim(1240,1240.5)
    ax2.plot(Pos,V)
    ax2.set_xlim(1240,1240.5)
    ax3.plot(Pos,I)
    ax3.set_xlim(1240,1240.5)
    
    
plt.show()
```
