```python
Energies=data[0,:,0]
Ir=data[0,:,4,None]/data[0,:,2,None]
for col in np.arange(4,12):
    Ir=np.concatenate((Ir,data[0,:,col,None]/data[0,:,2,None]),axis=1)   

fig=plt.figure(figsize=(8,4))

ax1=fig.add_subplot(211)
ax1.plot(Energies,Ir)
ax1=fig.add_subplot(212)
ax1.plot(Energies,np.mean(Ir,axis=1))
plt.show()
```
